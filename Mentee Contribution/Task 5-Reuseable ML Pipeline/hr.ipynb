## IMPORT LIBRARIES

```python
import numpy as np
import pandas as pd
import sklearn
```

## Load Dataset

```python
df=pd.read_csv("hr.csv")
df.head()
```

```python
df.columns
```

## EDA

```python
df.info()
```

```python
df.duplicated()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['Attrition'] = le.fit_transform(df['Attrition'])
```

```python
x = df.drop('Attrition', axis=1)
y = df['Attrition']
```

```python
x = x.drop(columns=[
    'EmployeeCount',
    'EmployeeNumber',
    'Over18',
    'StandardHours'
])
```

## **STEP 1** *DIVIDING THE DATA INTO INPUT AND OUTPUT*

```python
x=df.drop(['Attrition'],axis=1)
```

```python
x
```

```python
y=df[['Attrition']]
```

```python
y
```

## **STEP 2** *SPLIT THE DATA INTO TRAIN AND TEST*

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x,y,train_size=0.8,random_state=3)
```

```python
x_train
```

```python
x_train.reset_index(inplace=True)
x_train
```

```python
x_test.reset_index(inplace=True)
x_test
```

```python
y_train
```

```python
y_test
```

## **STEP 3** *FEATURE ENGINEERING ON TRAINING DATA*

```python
cat_df=x_train.select_dtypes(include='object')
cat_df
```

```python
from sklearn.preprocessing import OneHotEncoder
OHE=OneHotEncoder(sparse_output=False).set_output(transform='pandas')
cat_df_trans=OHE.fit_transform(cat_df)
```

```python
cat_df_trans
```

```python
OHE.categories_
```

```python
num_df = x_train[['Age', 'DailyRate', 'DistanceFromHome', 'Education', 'EnvironmentSatisfaction', 'HourlyRate', 'JobInvolvement', 'JobLevel', 'JobSatisfaction', 'MonthlyIncome', 'MonthlyRate', 'NumCompaniesWorked', 'PercentSalaryHike', 'PerformanceRating', 'RelationshipSatisfaction', 'StockOptionLevel', 'TotalWorkingYears', 'TrainingTimesLastYear', 'WorkLifeBalance', 'YearsAtCompany', 'YearsInCurrentRole', 'YearsSinceLastPromotion', 'YearsWithCurrManager']]
num_df
```

```python
no_trans = x_train[['PerformanceRating', 'StockOptionLevel']]
no_trans
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler().set_output(transform='pandas')
num_df_trans=scaler.fit_transform(num_df)
```

```python
num_df_trans
```

```python
num_df_trans.reset_index(inplace=True)
```

```python
cat_df_trans.reset_index(inplace=True)
```

```python
no_trans.reset_index(inplace=True)
```

```python
x_train_trans=pd.concat([num_df_trans, cat_df_trans, no_trans], axis=1)
```

```python
x_train_trans
```

```python
x_train_trans.pop('index')
```

```python
x_train_trans
```

## **STEP 4** *TRAINING MODEL USING ALGORITHM*

```python
from sklearn.neighbors import KNeighborsClassifier

model=KNeighborsClassifier()
# x_train_trans=df.drop(['Gender'],axis=1)
model.fit(x_train_trans, y_train)
```

## step5 - *Feature Transformations on Test Data*

```python
x_test
```

```python
num_df=x_test[['Age', 'DailyRate', 'DistanceFromHome', 'Education', 'EnvironmentSatisfaction', 'HourlyRate', 'JobInvolvement', 'JobLevel', 'JobSatisfaction', 'MonthlyIncome', 'MonthlyRate', 'NumCompaniesWorked', 'PercentSalaryHike', 'PerformanceRating', 'RelationshipSatisfaction', 'StockOptionLevel', 'TotalWorkingYears', 'TrainingTimesLastYear', 'WorkLifeBalance', 'YearsAtCompany', 'YearsInCurrentRole', 'YearsSinceLastPromotion', 'YearsWithCurrManager']]
num_df
```

```python
cat_df = x_test.select_dtypes(include='object')
```

```python
no_trans= x_test[['PerformanceRating', 'StockOptionLevel']]
```

```python
# num trans
num_df_trans=scaler.transform(num_df)
num_df_trans
```

```python
# cat df transformation

cat_df_trans=OHE.transform(cat_df)
cat_df_trans
```

```python
num_df_trans.reset_index(inplace=True)
```

```python
cat_df_trans.reset_index(inplace=True)
```

```python
no_trans.reset_index(inplace=True)
```

```python
x_test_trans=pd.concat([num_df_trans, cat_df_trans, no_trans],axis=1)
```

```python
x_test_trans.pop('index')
```

```python
x_test_trans
```

## step6 - *Predicting on x_test_Trans*

```python
y_pred=model.predict(x_test_trans)
```

```python
y_pred
```

## step7 - passing y_test(actual) and y_pred(model_prediction) to evaluation metrics

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

## *I built a reusable HR employee attrition prediction pipeline using scikit-learn. The pipeline preprocesses the employee data by handling missing values, encoding categorical features, and scaling numerical features. The preprocessed data is saved as hr.pkl to ensure the same transformations are applied to new data. The pipeline then trains a machine learning model, evaluates its performance using appropriate metrics, and saves the trained model for future predictions. This workflow can be reused to predict whether a new employee is likely to leave the organization without manually repeating the preprocessing and model training steps.*
